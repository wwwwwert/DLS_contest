```python
import numpy as np
```

```python
def func(x):
    return x ** 4

def df_dx(x):
    return 4 * x ** 3

def grad_descent(alpha = 0.1):
    min_f = 0
    prev_dist_min = 10000
    x = 4
    dist_mins = []

    for i in range(10000):
        dist_mins.append(func(x))
        x = x - alpha * df_dx(x)
    sns.lineplot(y=dist_mins, x=np.arange(len(dist_mins)))
    print(dist_mins[-4:])
```

```python
grad_descent(1 / 32 - 0.000001)
```

```python
grad_descent(1 / 32)
```

```python
grad_descent(1 / 32 + 0.000001)
```

Получается, что при $\alpha > \frac{1}{32}$ градиентный спуск не сойдётся.
